```python
nyc_jobs['test'] = nyc_jobs['To Apply'].astype('str')
```

```python
def text_process(mess):
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    stops = set(stopwords.words('english'))
    list_of_words = [word.lower() for word in nopunc.split() if word.lower() not in stopwords.words('english')] 
    return ' '.join(list_of_words)
```

```python
nyc_jobs['apply change'] = nyc_jobs['test'].apply(text_process)
```

```python
nyc_jobs['apply change'].dropna().apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).sort_values(ascending=False)
```

```python
nyc_jobs['skills change'] = nyc_jobs['Preferred Skills'].apply(text_process)
```

```python
x_temp = nyc_jobs['skills change'].dropna().apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).sort_values(ascending=False)
```

```python
top_words = pd.DataFrame(x_temp).reset_index()
```

```python
top_words.columns = ['word', 'word_freq']
```
