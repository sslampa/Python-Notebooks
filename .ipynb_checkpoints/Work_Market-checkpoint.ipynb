# 1) Who has the most openings?

To answer this question, I will explore this question by looking at the <strong># Of Positions</strong> by <strong>Agency</strong> and the year of the <strong>Posting Date</strong>.

To start I will sum all the <strong># Of Positions</strong> by <strong>Agency</strong>.

```python
agency_num_pos = pd.DataFrame(nyc_jobs.groupby('Agency')['# Of Positions'].sum()).reset_index()
```

Here is a quick visualization of the number of open positions by <strong>Agency</strong>. To make it cleaner I only used the top 20 in open positions.

```python
agency_num_pos.sort_values('# Of Positions', ascending=False, inplace=True)
```

```python
# Plots the top 20 open positions
agency_num_pos.head(20).plot('Agency', '# Of Positions', kind='barh', figsize=(12,6), color = 'red', alpha=.75)
```

Let's take a look at the table to get a better look of each company. We can see that <strong>DEPT OF HEALTH/MENTAL HYGIENE</strong> has a large amount of openings compared to the rest of the agencies. After the top six agencies in open positions, there is a large drop-off and the levels begin to even out.

```python
agency_num_pos.head(20)
```

Due to the drop-off, I will be using the top 6 agencies and get a closer look at how their number of openings change over the years.

```python
# Turn Posting Date into datetime object, then extract year
nyc_jobs['Posting Date'] = pd.to_datetime(nyc_jobs['Posting Date'])
nyc_jobs['Posting Date Year'] = pd.DatetimeIndex(nyc_jobs['Posting Date']).year
nyc_jobs['Posting Date Month'] = pd.DatetimeIndex(nyc_jobs['Posting Date']).month
```

```python
# Sum of all # Of Positions by Posting Date Year
all_agency_yr = pd.DataFrame(nyc_jobs.groupby(['Posting Date Year'])['# Of Positions'].sum()).reset_index()
```

```python
# Create a dataframe for top six open positions by agency and posting date year 
top_six = set(agency_num_pos['Agency'][:6].values)
top_six_df = nyc_jobs[nyc_jobs['Agency'].isin(top_six)]
top_six_agency_yr = pd.DataFrame(top_six_df.groupby(['Agency', 'Posting Date Year'])['# Of Positions'].sum()).reset_index()
```

```python
# Plot the bar plot of all all agencies summed
plt.figure()
ax = all_agency_yr.plot('Posting Date Year', '# Of Positions', kind='bar', 
                        figsize=(16,8), alpha=.25, legend=False)
ax1 = ax.twiny()

# Plots individual lines for each top six agency
for x in top_six:
    top_six_agency_yr[top_six_agency_yr['Agency'] == x].plot('Posting Date Year', '# Of Positions', 
                                                             kind='line', ax=ax1, linewidth=5,
                                                            label=x)

ax1.legend(loc='upper left')
plt.show()
```

From the above plot we see that an overwhelmingly amount of open positions come from 2015. This makes sense as the dataset contains job postings as of February 23, 2016. Let's get break the plot down further by looking at the amount of job postings by month and year.

```python
all_agency_month_yr = pd.DataFrame(nyc_jobs.groupby(['Posting Date Year', 'Posting Date Month'])
                                   ['# Of Positions'].sum()).reset_index()
```

```python
top_six_agency_month_yr = pd.DataFrame(top_six_df.groupby(['Agency', 'Posting Date Year', 'Posting Date Month'])
                                       ['# Of Positions'].sum()).reset_index()
```

```python
# Plot the bar plot of all all agencies summed for 2015 year
plt.figure()
ax = all_agency_month_yr[all_agency_month_yr['Posting Date Year'] == 2015].plot('Posting Date Month', 
                                                                                '# Of Positions', kind='bar',
                                                                                figsize=(16,8), alpha=.25, legend=False)
ax1 = ax.twiny()

# Plots individual lines for each top six agency
for x in top_six:
    top_six_agency_month_yr[(top_six_agency_month_yr['Agency'] == x) & \
                            (top_six_agency_month_yr['Posting Date Year'] == 2015)].plot('Posting Date Month', '# Of Positions', 
                                                             kind='line', ax=ax1, linewidth=5,
                                                            label=x)

ax1.legend(loc='upper left')
ax.set_ylim([0,1600])
ax1.set_ylim([0,1600])
plt.show()

# Plot the bar plot of all all agencies summed for 2016 year
plt.figure()
ax = all_agency_month_yr[all_agency_month_yr['Posting Date Year'] == 2016].plot('Posting Date Month', 
                                                                                '# Of Positions', kind='bar', 
                                                                                figsize=(16,8), alpha=.25, legend=False)
ax1 = ax.twiny()

# Plots individual lines for each top six agency
for x in top_six:
    top_six_agency_month_yr[(top_six_agency_month_yr['Agency'] == x) & \
                            (top_six_agency_month_yr['Posting Date Year'] == 2016)].plot('Posting Date Month', '# Of Positions', 
                                                             kind='line', ax=ax1, linewidth=5,
                                                            label=x)

ax1.legend(loc='upper left')
plt.show()
```

The above plots shows the obvious that, generally, more positions are open closer to the processing date of the dataset. <strong>DEPT OF HEALTH/MENTAL HYGIENE</strong> consistently have the most open positions. Let's explore the dataset further by looking to see which agency has the most unique job postings. A lot of the job postings show that they post both internally and externally for the same position. It is possible that agencies could be posting the same job over and over to get more applicants.

```python
# Group by Civil Service Title to find some of unique values
agency_unique_titles = pd.DataFrame(nyc_jobs.groupby('Agency')
                                    ['Civil Service Title'].unique().apply(lambda x: len(x))).reset_index()
```

```python
# Table showing top 10 Agencies with sum of unique Civil Service Title
agency_unique_titles.sort_values('Civil Service Title', ascending=False).head(10)
```

Even when looking at the unique job postings, <strong>DEPT OF HEALTH/MENTAL HYGIENE</strong> still have the most openings. Based on the above table data and the previous plots, it is safe to say that the <strong>DEPT OF HEALTH/MENTAL HYGIENE</strong> have the most openings.

# 2) Which departments have the highest and the lowest paying positions (based on the current job openings)?

First we need to define current job openings. From the above plot showing the openings for each month, there is a noticeably large increase in open postings starting in September 2015. As such, we will only use postings from the date to the process date of February 23, 2016.

```python
# Create current jobs dataframe 
current_jobs_df = nyc_jobs[nyc_jobs['Posting Date'] >= '2015-09-01'].reset_index(drop=True)
```

There are three features (<strong>Salary Range From</strong>, <strong>Salary Range To</strong>, and <strong>Salary Frequency</strong>) that specifically deal with salary. The first two of those deal with the minimum and maximum values offered for the position and the <strong>Salary Frequency</strong> is how often an employee will be paid. Let's explore each of the salary frequencies to answer the question.

```python
# Show unique values for Salary Frequency and value counts
freq_unique = set(current_jobs_df['Salary Frequency'].unique())
freq_values = current_jobs_df['Salary Frequency'].value_counts()
print freq_unique
print freq_values
```

As we see above, there are three unique values: <em>Annual</em>, <em>Daily</em>, and <em>Hourly</em>. Let's get a better look at each value by plotting histograms. The dark blue represents the <strong>Salary Range From</strong>(or minimum) and the light green represents the <strong>Salary Range To</strong>(or maximum) salary to be paid.

```python
# Create histogram showing minimum and maximum salary 
ax = current_jobs_df[current_jobs_df['Salary Frequency'] == 'Annual'].hist('Salary Range From', range=(25000, 225000))
current_jobs_df[current_jobs_df['Salary Frequency'] == 'Annual'].hist('Salary Range To', ax=ax, alpha=.5, range=(25000, 225000))
plt.title('Annual Salary')
```

```python
# Histogram of hourly salary frequency 
ax1 = current_jobs_df[current_jobs_df['Salary Frequency'] == 'Hourly'].hist('Salary Range From', range=(10, 100))
current_jobs_df[current_jobs_df['Salary Frequency'] == 'Hourly'].hist('Salary Range To', ax=ax1, alpha=.5, range=(10, 100))
plt.title('Hourly Salary')
```

```python
# Histogram of daily salary frequency
ax2 = current_jobs_df[current_jobs_df['Salary Frequency'] == 'Daily'].hist('Salary Range From', bins=8, range=(200, 450))
current_jobs_df[current_jobs_df['Salary Frequency'] == 'Daily'].hist('Salary Range To', ax=ax2, alpha=.5, bins=8, range=(200, 450))
plt.title('Daily Salary')
```

The above 3 plots show the minimum(dark blue) and maximum(light green) salaries offered. We see for the <em>Annual</em> and <em>Hourly</em> show a right skew. The <em>Daily</em> shows little information due to the low amount of jobs offering that <strong>Salary Frequency</strong>. Going forward, <em>Daily</em> will not be considered.

 We will use the average of <strong>Salary Range From</strong> and <strong>Salary Range To</strong> to determine the lowest and highest paying positions.

```python
# Create new column for average salary
current_jobs_df['Salary Average'] = (current_jobs_df['Salary Range From'] + current_jobs_df['Salary Range To']) / 2.0
```

With the average in place, let's group by <strong>Salary Frequency</strong> for <em>Annual</em> and <em>Hourly</em> and rank each position by its <strong>Salary Average</strong> percentile.

```python
# Create dataframe of percentile rank 
annual_rank = pd.DataFrame(current_jobs_df[current_jobs_df['Salary Frequency'] == 'Annual'].groupby \
                           ('Agency')['Salary Average'].rank(method='max', pct = 'True'))

# Rename column
annual_rank.columns = ['Rank']

# Left Join annual_rank and current_jobs_df
annual_rank = annual_rank.merge(current_jobs_df, left_index=True, right_index=True, how='left')
```

To keep all of the information and determine the highest and lowest paying departments, we will average the top and bottom 50%.

```python
# Split the ranks into high and low
annual_high = annual_rank[(annual_rank['Salary Frequency'] == 'Annual') & (annual_rank['Rank'] > .5)]
annual_low = annual_rank[(annual_rank['Salary Frequency'] == 'Annual') & (annual_rank['Rank'] <= .5)]

# Group by agency, then average
annual_high_mean = pd.DataFrame(annual_high.groupby('Agency')['Salary Average'].mean()).reset_index()
annual_low_mean = pd.DataFrame(annual_low.groupby('Agency')['Salary Average'].mean()).reset_index()

# Combine high and low, then change column names
annual_mean = annual_high_mean.merge(annual_low_mean, on='Agency')
annual_mean.columns = ['Agency', 'High', 'Low']
```

```python
# Scatter plot of High vs Low 
annual_mean.plot('Low', 'High', kind='scatter')
```

The above plot is High vs Low for <em>Annual</em>. We expect that points on the bottom-left are companies that have both low top and bottom end salaries. On the other side, we expect the top top-right to show companies that have high top and and bottom salaries.

There are two clear points that are at the top-right of the plot. <strong>DEPARTMENT OF SANITATION</strong> and <strong>NYC EMPLOYEES RETIREMENT SYS</strong> are the departments with the highest paying departments.

```python
annual_mean[(annual_mean['Low'] > 75000) & (annual_mean['High'] > 110000)]
```

<strong>DISTRICT ATTORNEY KINGS COUNTY</strong> is the clear lowest paying department for <em>Annual</em>.

```python
annual_mean[(annual_mean['Low'] < 40000) & (annual_mean['High'] < 55000)]
```

Let's now look at the data for <em>Hourly</em>. This will be the same process as <em>Annual</em>.

```python
# Split the ranks into high and low
hourly_rank = pd.DataFrame(current_jobs_df[current_jobs_df['Salary Frequency'] == 'Hourly'].groupby \
                           ('Agency')['Salary Average'].rank(method='max', pct = 'True'))

# Rename column 
hourly_rank.columns = ['Rank']

# Left join hourly_rank and current_jobs_df
hourly_rank = hourly_rank.merge(current_jobs_df, left_index=True, right_index=True, how='left')
```

```python
# Split the ranks into high and low
hourly_high = hourly_rank[(hourly_rank['Salary Frequency'] == 'Hourly') & (hourly_rank['Rank'] > .5)]
hourly_low = hourly_rank[(hourly_rank['Salary Frequency'] == 'Hourly') & (hourly_rank['Rank'] <= .5)]

# Group by agency, then average
hourly_high_mean = pd.DataFrame(hourly_high.groupby('Agency')['Salary Average'].mean()).reset_index()
hourly_low_mean = pd.DataFrame(hourly_low.groupby('Agency')['Salary Average'].mean()).reset_index()

# Combine both high and low, then rename columns
hourly_mean = hourly_high_mean.merge(hourly_low_mean, on='Agency')
hourly_mean.columns = ['Agency', 'High', 'Low']
```

```python
# Plot scatter plot High vs Low
hourly_mean.plot('Low', 'High', kind='scatter')
```

The same principles apply here where the top-right is the highest paying departments and the bottom-left are the lowest paying.

The <strong>FIRE DEPARTMENT</strong> is the far and away the highest paying hourly department.

```python
hourly_mean[(hourly_mean['Low'] > 45) & (hourly_mean['High'] > 70)]
```

There are four departments that have close points for the lowest paying departments: <strong>ADMIN TRIALS AND HEARINGS</strong>, <strong>DEPARTMENT OF BUSINESS SERV</strong>, <strong>DEPT OF CITYWIDE ADMIN SVCS</strong>, and <strong>OFFICE OF MANAGEMENT & BUDGET</strong>

```python
hourly_mean[(hourly_mean['Low'] < 15) & (hourly_mean['High'] < 30)]
```

# 3) Which jobs do you think are the hardest to fill? (What makes you say that?)

```python
mylist = ['no', 'not', 'none']
pattern = '|'.join(mylist)
res_change = nyc_jobs[~nyc_jobs['Residency Requirement'].str.contains(pattern)]
```

```python
somepattern = "([^ ]*) year|years"
res_change['exp year'] = res_change['Minimum Qual Requirements'].str.extract(somepattern)
```

```python
some_dict = {
    'one': '1',
    'two': '2',
    'three': '3',
    'four': '4',
    'five': '5',
    'six': '6',
    'seven': '7',
    'eight': '8',
    'nine': '9',
    'ten': '10',
    'one-half': '.5',
    '10+': '10',
    '4-6': '5'
}
```

```python
res_change['exp year'] = res_change['exp year'].str.lower().replace(some_dict)
res_change['exp year'] = res_change['exp year'].str.replace('\(|\)', '')

res_change.dropna(subset=['exp year'], inplace=True)
```

```python
exp_list = set(['2', '4', '3', '5', '6', '1', '.5', '7', '10', '8'])
```

```python
change = res_change[res_change['exp year'].isin(exp_list)]
```

```python
job_positions = pd.DataFrame(change.groupby('Civil Service Title')['# Of Positions'].sum()).reset_index()
```

```python
job_positions.hist(bins=25, range=(0,450))
```
